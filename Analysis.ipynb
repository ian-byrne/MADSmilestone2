```python
import numpy as np
import pandas as pd
import ast
from IPython.display import display
import imageio
import matplotlib.pyplot as plt

```

# Classification Report

```python
def show_class_report(mod_name, ext):
  path = '/content/gdrive/MyDrive/Colab Notebooks/model_charts/cr{}'.format(ext)
  cust_file = open(path, "r")
  #print(cust_file.readline())
  contents = cust_file.read()
  dictionary = ast.literal_eval(contents)
  cust_file.close()

  df = pd.DataFrame(dictionary)
  df.rename(columns = {'0': "Possible Dementia (0)", '1': "Likely Dementia (1)", '2': "No Dementia (2)"}, inplace = True)
  df.index.name = mod_name
  
  return df



mod_ext = [('Model 1', '.txt'), ('Model 2', '2.txt'), ('Model 3','3.txt'), ('Model 4', '4.txt'), 
           ('Model 4 fixed', '4_fix.txt'),('Model 5','5.txt'), ('Model 6','6.txt'), ('Model 7','7.txt'), 
           ('Model 8', '8.txt'), ('Model 9','9.txt'),  ('Model 10', '10.txt'), ('Resnet50 1', 'res.txt'), 
           ('Resent50 2', 'res2.txt'),('Resent50 3','res3.txt'), ('Resent50 4','res4.txt'), 
           ('Resent50 6', 'res6.txt'), ('Resent50 7', 'res7.txt')]

for ext in mod_ext:
  display(show_class_report(ext[0], ext[1]))
```

# Accuracy and Loss 

```python
def plot_acc_loss(ext):
  path = '/content/gdrive/MyDrive/Colab Notebooks/model_charts/acc_loss{}'.format(ext[1])
  im = imageio.imread(path)
  print(ext[0]+': \n')
  plt.figure(figsize=(10, 12))
  plt.imshow(im)
  plt.axis('off')
  plt.show()

mod_ext = [('Model 1', '.png'), ('Model 2', '2.png'), ('Model 3','3.png'), ('Model 4', '4.png'), 
           ('Model 4 fixed', '4_fix.png'),('Model 5','5.png'), ('Model 6','6.png'), ('Model 7','7.png'), 
           ('Model 8', '8.png'), ('Model 9','9.png'), ('Resnet50 1', 'res.png'), 
           ('Resnet50 2', 'res2.png'),('Resnet50 3','res3.png'), ('Resnet50 4','res4.png'), 
           ('Resnet50 6', 'res6.png'), ('Resnet50 7', 'res7.png')]
short_list = [('Model 1', '.png'), ('Model 2', '2.png'), ('Model 3','3.png'), ('Model 4 fixed', '4_fix.png'),
              ('Model 7','7.png'), ('Resnet50 1', 'res.png'),('Resnet50 2', 'res2.png'),
              ('Resnet50 3','res3.png'),('Resnet50 7', 'res7.png')]
for val in short_list:
  plot_acc_loss(val)
```
